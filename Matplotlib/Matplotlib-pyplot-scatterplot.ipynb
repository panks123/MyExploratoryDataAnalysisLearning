# Scatter plot
**Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another. Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them.**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[34,23,23,45,65,12,56,59,34,45]
y=[74,53,73,35,68,62,16,39,39,89]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
df=pd.read_csv("googleplaystore.csv",nrows=1000) # It reads only 1000 rows
df.shape
```

```python
df.head(7)
```

```python
x=df["Rating"]
y=df["Reviews"]

plt.scatter(x,y)
plt.xlabel('Rating')
plt.ylabel("Reviews")
plt.title("Rating vs Reviews of Playstore Apps")
plt.show()
```

```python
plt.scatter(x,y, c='r') # for applying red color
plt.xlabel('Rating')
plt.ylabel("Reviews")
plt.title("Rating vs Reviews of Playstore Apps")
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('Rating')
plt.ylabel("Reviews")
plt.title("Rating vs Reviews of Playstore Apps")
plt.show()
```

```python
plt.scatter(x,y,marker="s")
plt.xlabel('Rating')
plt.ylabel("Reviews")
plt.title("Rating vs Reviews of Playstore Apps")
plt.show()
```

```python
x=df["Rating"]
y=df["Installs"]

plt.scatter(x,y)
plt.xlabel('Rating')
plt.ylabel("Reviews")
plt.title("Rating vs Reviews of Playstore Apps")
plt.show()
```

**Scatter plot for multiple datasets**

```python
# dataset-1
x1 = [89, 43, 36, 36, 95, 10, 
      66, 34, 38, 20]
  
y1 = [21, 46, 3, 35, 67, 95, 
      53, 72, 58, 10]
  
# dataset2
x2 = [26, 29, 48, 64, 6, 5,
      36, 66, 72, 40]
  
y2 = [26, 34, 90, 33, 38, 
      20, 56, 2, 47, 15]
plt.figure(figsize=(12,7))
plt.scatter(x1,y1,c='r',marker='s')
plt.scatter(x2,y2,c='g',marker='^')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
