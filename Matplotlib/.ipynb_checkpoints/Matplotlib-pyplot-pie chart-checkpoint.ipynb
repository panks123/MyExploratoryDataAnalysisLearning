# Pie charts
**A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges. The area of the wedge is determined by the length of the arc of the wedge. The area of a wedge represents the relative percentage of that part with respect to whole data**

```python
import matplotlib.pyplot as plt
```

```python
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
```

```python
plt.pie(no_students)
plt.show()
```

```python
# So we have successfully plotted the pie chart of student data but it is not labelled
# let's label it
```

```python
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]

plt.pie(no_students,labels=lang)
plt.show()
```

```python
# Customizing colors

lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
colors=['r','g','b','c','m']
plt.pie(no_students,labels=lang,colors=colors)
plt.show()
```

```python
# 'explode' parameter - A wedge of a pie chart can be made to explode (i.e sliced) from the rest of the wedges of the pie chart
# explode parameter is passed as a list of values for each wedge

lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0.3,0,0]

plt.pie(no_students,labels=lang,explode=explode)
plt.show()
```

```python
#'autopct' parameter- It specifies the percentage space for each wedge
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0,0,0]

plt.pie(no_students,labels=lang,explode=explode,autopct='%0.2f%%') # 0.2 means two places after decimal
plt.show()
```

```python
# 'shadow' parameter It takes either True or False
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0.3,0,0]

plt.pie(no_students,labels=lang,explode=explode,shadow=True)
plt.show()
```

```python
# radius parameter
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0.03,0,0]

plt.pie(no_students,labels=lang,explode=explode,radius=2) # radius parameter custumizes the radius of the pie chart
plt.show()
```

```python
# start angle is the position from where the first wedge starts and goes in anticlockwise
# the default start angle is 0 i.e at the x-axis, but we can change the start angle by specifying a startangle parametyer
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0.03,0,0]

plt.pie(no_students,labels=lang,explode=explode,radius=1.4,startangle=90) 
plt.show()
```

```python
# textprops parameter it is used to custumize the text appearing on the pie chart (It is passed as a dictionary of parameters)
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0.03,0,0]

plt.pie(no_students,labels=lang,explode=explode,radius=1.4,textprops={'fontsize':25}) 
plt.show()
```

```python
# Custumizing wedge properties
lang=['Python','C','Java','R','C++']
no_students=[45,15,35,25,30]
explode=[0.1,0,0.03,0,0]

plt.pie(no_students,labels=lang,explode=explode,radius=1.4,textprops={'fontsize':15},
        wedgeprops={"linewidth":2,"edgecolor":'k'}) 
plt.legend(loc='center')
plt.show()
```
