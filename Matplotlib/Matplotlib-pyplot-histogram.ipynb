# What is histogram?
**Histogram is a graphical display of data using bars of different heights. A histogram is a display that indicates the frequency of specified ranges of continuous data values on a graph in the form of immediately adjacent bars.It is basically used for representing frequency distribution in forms of bars**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
group1_age=np.random.randint(18,45,100)
group1_age
```

```python
plt.hist(group1_age)
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

**So, the histogram is created by following steps:**
- Bin the range of values.
- Divide the entire range of values into a series of intervals.
- Count how many values fall into each interval

```python
# defining bin (i.e range of bars) explicitly
bins=[15,20,25,30,35,40,45,50]
```

```python
plt.hist(group1_age,bins=bins)
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

```python
# We can see that all the bars are collapsed, we can provide the relative width between the bars of different bins
# by the parameter rwidth-The relative width of the bars as a fraction of the bin width
plt.hist(group1_age,bins=bins,rwidth=0.9)
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.hist(group1_age,bins=bins,rwidth=0.5) # on decreasing the rwidth value separation between the bars increases
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.hist(group1_age,bins=bins,rwidth=0.5,color='g') # bars appears in green color
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.hist(group1_age,bins=bins,histtype='step') 
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.hist(group1_age,bins=bins)
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.hist(group1_age,bins=bins,rwidth=0.5,color='g',orientation='horizontal') # by default orientation is horizontal
plt.title("Age distribution")
plt.ylabel("Group1 age")
plt.xlabel("No. of students")
plt.legend(['Group1'])
plt.show()
```

```python
group2_age=np.random.randint(15,40,100)
group2_age
```

```python
# Plotting for multiple groups
plt.hist(group1_age,bins=bins,rwidth=0.9,color='g') 
plt.hist(group2_age,bins=bins,rwidth=0.7,color='r') 
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.legend(['Group1','Group2'])
plt.show()
```

```python
plt.hist(group1_age,bins=bins,rwidth=0.9,color='g') 
plt.hist(group2_age,bins=bins,rwidth=0.7,color='r',orientation='horizontal') 
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.legend(['Group1','Group2'])
plt.show()
```

```python
# best way to plot histogrms of multiple groups
plt.hist([group1_age,group2_age],bins=bins,rwidth=0.9,color=['g','m'])  # just pass them as a list in same function
plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.legend(['Group1','Group2'])
plt.show()
```

```python
# to change the size of the histogram
plt.figure(figsize=(16,9)) # It should be the first line
plt.hist([group1_age,group2_age],bins=bins,rwidth=0.9,color=['g','m'])  # just pass them as a list in same function

plt.title("Age distribution")
plt.xlabel("Group1 age")
plt.ylabel("No. of students")
plt.legend(['Group1','Group2'])
plt.show()
```
