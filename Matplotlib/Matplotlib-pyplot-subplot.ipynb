# Subplot
**to create a figure and a set of subplots**

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1) # (nrow,ncol,index)It means the plot is divided into a grid of two rows and two cols and we are plotting the first subplot
plt.subplot(2,2,3)
plt.subplot(2,4,2)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.pie([1,3,1],labels=['One','Two','Three'])

plt.subplot(2,2,2)
plt.pie([11,31,14],labels=['One','Two','Three'])

plt.subplot(2,2,3)
plt.pie([101,35,100],labels=['One','Two','Three'])

plt.subplot(2,2,4)
plt.pie([41,31,14],labels=['One','Two','Three'])

plt.show()
```

```python
import numpy as np
x=np.arange(-100,100)
y1=np.sin(x*100)
y2=np.cos(x*100)
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2,1,1)
plt.plot(x,y1,color='r')
plt.xlabel('x')
plt.ylabel('sin(x*100)')
plt.title('y1=np.sin(x*100)')
plt.subplot(2,1,2)
plt.plot(x,y2,color='g')
plt.xlabel('x')
plt.ylabel('cos(x*100)')
plt.title('y2=np.cos(x*100)')
plt.show()
```
