# Bar plot
**A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.**

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
```

```python
lang=["Python","R","Java","Scala","Julia"]
class1=[30,30,40,25,43]
class2=[50,34,43,12,30]
class3=[67,45,23,34,23]
```

```python
plt.bar(lang,class1)
plt.show()
```

```python
# to plot the bar graph horizontally
plt.barh(lang,class1)
plt.show()
```

```python
# to specify the width of the bars
plt.bar(lang,class1,width=0.2)
plt.show()
```

```python
plt.bar(lang,class1,width=0.2,color='g',align='edge')
plt.show()
```

```python
plt.bar(lang,class1,width=0.2,color='g',alpha=.3) # changes the opacity of the color
plt.show()
```

```python
plt.bar(lang,class1,width=0.4,edgecolor='r',linestyle='--',linewidth=3)
plt.title("Class1-Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.show()
```

```python
# to hide the plot
plt.bar(lang,class1,width=0.2,visible=False)
plt.show()
```

```python
style.use("ggplot")
plt.bar(lang,class1,width=0.4,edgecolor='r',linestyle='--',linewidth=3)
plt.title("Class1-Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.show()
```

```python
plt.bar(lang,class1,width=0.4,color='r')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.show()
```

**plotting bar charts for multiple groups-Multiple bar charts are generally used for comparing different entities.**

```python
plt.bar(lang,class1,width=0.4,color='r')
plt.bar(lang,class2,width=0.5,color='g')
plt.bar(lang,class3,width=0.3,color='b')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.legend(['Class1','Class2','Class3'])
plt.show()
```

```python
# we can see that the bar charts are overlapping / stacked  so we need to follow other method for this
```

```python
X_axis = np.arange(len(lang))
width=0.3
plt.bar(X_axis,class1,width=width,color='r')
plt.bar(X_axis+width,class2,width=width,color='g')
plt.bar(X_axis+width+width,class3,width=width,color='b')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.legend(['Class1','Class2','Class3'])
plt.show()
```

```python
# Above we can see that X axis ticks are values but we want the language names
X_axis = np.arange(len(lang))
width=0.3
plt.bar(X_axis,class1,width=width,color='r')
plt.bar(X_axis+width,class2,width=width,color='g')
plt.bar(X_axis+width+width,class3,width=width,color='b')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
# for that
plt.xticks(X_axis, lang) # This will replace the X_axis values by lang values (i.e lang. names ) in X - axis
plt.legend(['Class1','Class2','Class3'])
plt.show()
```

```python
# Changing the position of xticks
X_axis = np.arange(len(lang))
width=0.3
plt.bar(X_axis,class1,width=width,color='r')
plt.bar(X_axis+width,class2,width=width,color='g')
plt.bar(X_axis+width+width,class3,width=width,color='b')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.xticks(X_axis+width, lang) # This will replace the X_axis+ width values by lang values (i.e lang. names ) in X - axis
plt.legend(['Class1','Class2','Class3'])
plt.show()
```

```python
# Rotating the xtickss
X_axis = np.arange(len(lang))
width=0.3
plt.bar(X_axis,class1,width=width,color='r')
plt.bar(X_axis+width,class2,width=width,color='g')
plt.bar(X_axis+width+width,class3,width=width,color='b')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.xticks(X_axis+width, lang,rotation=45) # this will rotate the xticks by 45 degrees 
plt.legend(['Class1','Class2','Class3'])
plt.show()
```

```python
X_axis = np.arange(len(lang))
width=0.3
plt.bar(X_axis,class1,width=width,color='r')
plt.bar(X_axis+width,class2,width=width,color='g')
plt.bar(X_axis+width+width,class3,width=width,color='b')
plt.title("Subject wise no. of students")
plt.xlabel("Subjects")
plt.ylabel("No. of students")
plt.xticks(X_axis+width, lang,rotation=90) # this will rotate the xticks by 90 degrees 
plt.legend(['Class1','Class2','Class3'])
plt.show()
```
