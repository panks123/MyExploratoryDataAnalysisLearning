# Seaborn jointplot

**Seaborn's jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins. It displays relationship between two variables (bivariate) as weell as the distrbution of the individual variables in the margins (univariate)**

- Join distribution estabilishes relationship between two variables
- Marginal distribtion shows the distribution of the individual variables
- combining these two we get the  joint plot as a whole

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('geyser')
```

```python
df.head()
```

```python
df.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x='waiting',y='duration',data=df)
plt.show()
```

# we can plot different types of joint plots:
- sctter (default)
- kde
- reg
- hist
- hex
- resid

```python
# kde - kernel density estimation plot

sns.jointplot(x='waiting',y='duration',data=df, kind='kde')
plt.show()
```

```python
# reg plot 
sns.jointplot(x='waiting',y='duration',data=df, kind='reg')
plt.show()
# here the central plot has both scattered plot as well as the regression line and 
# the marginal plots have histogram as well as the kde plot
```

```python
# hist plot

sns.jointplot(x='waiting',y='duration',data=df, kind='hist')
plt.show()
```

```python
# drwaing inference with categorical data using 'hue' parameter

sns.jointplot(x='waiting',y='duration',data=df, kind='scatter',hue='kind')
plt.show()
```

```python
sns.jointplot(x='waiting',y='duration',data=df, kind='kde',hue='kind')

plt.show()
```

```python
# Changing the color using 'color' parmeter
sns.jointplot(x='waiting',y='duration',data=df,color='g')
plt.show()
```

```python
# But if we are already using the 'hue' parameter then in that case 'palette' parameter is used

sns.jointplot(x='waiting',y='duration',data=df, kind='scatter',hue='kind',palette='autumn')
plt.show()
```

```python
# to style the main plot separately we use 'joint_kws' parameter which is passed as a dictionary

sns.jointplot(x='waiting',y='duration',data=df, kind='scatter',joint_kws={'marker':'^'})
plt.show()
```

```python
# If we don't want the shaded regression line then
sns.jointplot(x='waiting',y='duration',data=df, kind='reg',joint_kws={'ci':None})
plt.show()
```

```python
# to style the marginal plot separately we use 'marginal_kws' parameter which is passed as a dictionary

sns.jointplot(x='waiting',y='duration',data=df, kind='scatter',marginal_kws={'color':'r'})
plt.show()
```
