# Countplot
**Show the counts of observations in each categorical bin using bars. A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.n  It is very useful while doing exploratory data analysis (EDA)**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.sex.value_counts()
```

```python
sns.countplot(x='sex',data=df);
```

```python
df.day.value_counts()
```

```python
sns.countplot(x='day',data=df);
```

```python
sns.countplot(y='day',data=df);
```

```python
# Sorting the bars in descending order
```

```python
df.day.value_counts() # We can see that value_counts returns sorted categories according to their counts values
```

```python
# So
df.day.value_counts().index 
```

```python
sns.countplot(x='day',data=df,order=df.day.value_counts().index); # In descending order
```

```python
sns.countplot(x='day',data=df,order=df.day.value_counts().index[::-1]); # In ascending order
```

```python
# hue parameter
sns.countplot(x='day',data=df,order=df.day.value_counts().index,hue='sex');
```

```python
# hue_order parameter
sns.countplot(x='day',data=df,order=df.day.value_counts().index,hue='sex',hue_order=['Male','Female']); # It is by default
```

```python
sns.countplot(x='day',data=df,order=df.day.value_counts().index,hue='sex',hue_order=['Female','Male']);
```

```python
# Styling the count plot
```

```python
# color parameter -- It will specify only one color and all the bars will appear in same color
```

```python
sns.countplot(x='day',data=df,color='g');
```

```python
# palette parameter -- for different colored bars

sns.countplot(x='day',data=df,palette='autumn');
```

```python
# lw (linewidth) and ec (edge color)
sns.countplot(x='day',data=df,lw=3,ec='k');
```

```python
# hatch
sns.countplot(x='day',data=df,hatch='*');
```

```python
sns.countplot(x='day',data=df,hatch='/',lw=3,ec='k');
```
