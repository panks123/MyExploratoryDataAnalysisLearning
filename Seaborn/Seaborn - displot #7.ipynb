# Seaborn displot

**Distribution plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('mpg').dropna()
df.head()
```

```python
sns.displot(df['weight']);
```

```python
sns.displot(df['weight'],bins=20);
```

```python
# Including kde plot
sns.displot(df['weight'],kde=True,lw=5);
```

```python
# Including rugplot

sns.displot(df['weight'],rug=True,rug_kws={'color':'r','height':0.15});
```

```python
# kind parameter 

sns.displot(df['weight'],kind='kde');
```

```python
sns.displot(df['weight'],kind='kde',fill=True,lw=5);
```

```python
sns.displot(df['weight'],kind='hist');
```

```python
sns.displot(df['weight'],kind='ecdf'); # ecdf plot apllies empirical cumulative distribution function -> 
#                                        this adds all the values of weights as it moves along x-axis (cumulative sum)
```

```python
# hue parameter -> including categorical variable in the displot

sns.displot(x='weight',data=df,hue=df['origin']);
```

```python
sns.displot(x='weight',data=df,kind='kde',hue=df['origin']);
```

```python
# Till now we have done 1D univariate displot

# Now we will do 2D i.e. bivariate displot
```

```python
sns.displot(x='weight',y='horsepower',data=df)# wherever the color is more dense that represent distribution with higher density
plt.show()
```

```python
sns.displot(x='weight',y='horsepower',data=df,kind='kde',color='r')
plt.show()
```
