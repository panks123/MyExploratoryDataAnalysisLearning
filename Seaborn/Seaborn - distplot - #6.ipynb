# Seaborn distplot

**It is the distribution plot. It works with numerical data.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('mpg').dropna()
df.head()
```

```python
sns.distplot(df['weight']);
plt.show()
```

```python
# So we can see that in the distplot by default the histogram and kde plot are plotted

# Let's remove the kde plot

sns.distplot(df['weight'], kde=False);
plt.show()
```

```python
# We can also turn off the histogram

sns.distplot(df['weight'],hist=False);
plt.show()
```

```python
# We can control bins

sns.distplot(df['weight'],bins=20); # 20 bins histogram
plt.show()
```

```python
sns.distplot(df['weight'],bins=5); # 5 bins histogram
plt.show()
```

```python
sns.distplot(df['weight'],bins=[2000,2500,3500,4000,5200]); # Bins that we want
plt.show()
```

```python
# styling the histogram by 'hist_kws' argument

sns.distplot(df['weight'],hist_kws={'color':'g','alpha':1});
plt.show()
```

```python
sns.distplot(df['weight'],hist_kws={'color':'g','alpha':1,'histtype':'step'});
plt.show()
```

```python
# Styling the kde plot using 'kde_kws' parameter

sns.distplot(df['weight'],hist_kws={'color':'g','alpha':1,'histtype':'step'},kde_kws={'color':'r','lw':5});
plt.show()
```

```python
# Rug plot
sns.distplot(df['weight'],rug=True,rug_kws={'color':'y'});
plt.show()
```

```python
sns.distplot(df['weight'],rug=True,rug_kws={'color':'y','height':0.25});
plt.show()
```
