# Line plot using seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[12,45,23,54,34,56,54,23,33,54]
y=[43,31,23,52,24,26,34,53,53,54]
plt.show()
```

```python
sns.relplot(x,y,kind='line')
plt.show()
```

```python
sns.lineplot(x,y)
plt.show()
```

```python
sales=pd.read_excel('US Superstore data.xls')
sales.head()
```

```python
sns.relplot('Order Date','Sales',data=sales,kind='line')
plt.show()
```

```python
sns.lineplot('Order Date','Sales',data=sales)
plt.show()
```

```python
sales['year']=sales['Order Date'].dt.year
sales.head(3)
```

```python
sales['year'].unique()
```

```python
# We can derive inference that how  sale varies over years
sns.relplot('year','Sales',data=sales,kind='line')
plt.xticks([2016, 2015, 2014, 2017])
plt.show()
```

```python
# The blue shade shown in the graph shows the confidence interval 
# We can reove it as
sns.relplot('year','Sales',data=sales,kind='line',ci=None)
plt.xticks([2016, 2015, 2014, 2017])
plt.show()
```

```python
# We can derive inference that how the different Segment sale varies over years
sns.lineplot('year','Sales',data=sales,hue='Segment')
plt.xticks([2016, 2015, 2014, 2017])
plt.show()
```

```python
# We can derive inference that how the different Categories sale varies over years
sns.lineplot('year','Sales',data=sales,hue='Category')
plt.xticks([2016, 2015, 2014, 2017])
plt.show()
```

```python

sns.lineplot('year','Sales',data=sales,hue='Ship Mode',style='Category')
plt.xticks([2016, 2015, 2014, 2017])
plt.show()
```

```python

sns.relplot('year','Sales',data=sales,col='Category',kind='line')
plt.xticks([2016, 2015, 2014, 2017])
plt.show()
```
