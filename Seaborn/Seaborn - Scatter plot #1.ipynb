# Seaborn library
**Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps you explore and understand your data.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Scatter plot or relation plot using seaborn

```python
sales=pd.read_excel('US Superstore data.xls')
sales.head()
```

```python
sales.shape
```

```python
sns.relplot(x='Sales',y='Profit',data=sales)
plt.show()
```

```python
# Inference from the scatter plot -- In some cases profit is below 0 i.e loss is thereand in some cases profit is higher and 
# some outlier points are also there
```

```python
sales['Segment'].value_counts()
```

```python
# Let us color code our scatter plot based on the categorical variale 'Segment'

sns.relplot(x='Sales',y='Profit',data=sales,hue='Segment')
plt.show()
```

```python
sales['Category'].value_counts()
```

```python
# based on the categorical variale 'Segment'
sns.relplot(x='Sales',y='Profit',data=sales,hue='Category')
plt.show()
```

```python
# Custumizing markers
sns.relplot(x='Sales',y='Profit',data=sales,hue='Category',style='Category')
plt.show()
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,hue='Category',style='Segment')
plt.show()
```

```python
# size parameter - will produce elements with different sizes
# sizes : It is passed as list, dict, or tuple
#         It determines how sizes are chosen when ``size`` is used.
sns.relplot(x='Sales',y='Profit',data=sales,hue='Category',size='Discount',sizes=(30,200))
plt.show()
```
