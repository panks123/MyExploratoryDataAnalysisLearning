# Seaborn boxplot

**Boxplot captures the summary of the data efficiently with a simple box and whiskers and allows us to compare easily across groups. Boxplots summarizes a sample data using 25th, 50th and 75th percentiles.The advantage of comparing quartiles is that they are not influenced by outliers.**

```python
# 1st the median(Q2), 25% (Q1), and 75% (Q3) quartiles are calculated and represented as lines , these lines are joined at 
# their ends to form a box ,  
```

```python
# The other part of the box plot is the whisker - for finding the whisker we first measure the Inner Quartile Range
# The distance between the 25% and the 75% line in the boxplot is called Inner Quartile Range
```

```python
# Once we know the IQR we extend the Whiskers in either direction of the Box by 1.5 times IQR, the data points that lie outside
# the 1.5*IQR range are called outliers and for the outlier a diamond point is represented
```

- Extreme left black vertical line the box plot is the minimum value
- First black vertical line of rectangle shape of box plot is First quartile or 25%
- Second black vertical line (i.e. middle one)of rectangle shape of the box plot is Second quartile or 50% or median.
- Third black vertical line of rectangle shape of the box plot is third quartile or 75%
- Top black vertical line of rectangle shape of the box plot is maximum value.
- Small diamond shape of the box plot is outlier data or erroneous data.

```python
# So box plot helps us to visualize the  statistical summary and it also helps to find if the data has some outliers or not
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
cars=sns.load_dataset('mpg').dropna()
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars['cylinders'].value_counts()
```

```python
# Filtering cars with only even number of cylinders
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
cars['cylinders'].value_counts()
```

```python
sns.boxplot(x=cars['mpg']);
```

```python
cars['mpg'].describe()
```

```python
# min        9.000000
# 25%       17.000000
# 50%       23.000000
# 75%       29.000000
# max       46.600000

# We can clearly see these data plotted in the box plot
```

```python
# Plotting the box plot based on groups

sns.boxplot(x=cars.origin,y=cars['mpg']);
```

```python
# So seaborn groups the data based on the category -> x='origin' and for each group it 
# does the box plot calculations for each of those groups based on y='mpg' (miles per gallon)
```

```python
sns.boxplot(x='origin',y='mpg',data=cars); # Same as above 
```

```python
# hue parameter
# We can split the data into groups not only by one category but also by two categories using 'hue parameter'
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,hue='cylinders');
```

```python
# In the earler box plot we see that USA has least values for 'mpg' but now we can see the rason behind it that USA produces 
# all types of cylinder cars (4,6,8)
```

```python
# Let's create our own categorical variable

cars.model_year.describe()
```

```python
# We can see that most no. of cars are manufactured inthe year > 1976

# So we create the new categorical variable as

cars['newer_model']=cars.model_year>76
```

```python
cars.head()
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model');
```

```python
# So we can clearly visualize that the newer model cars( those which were manufactured in the years > 1976 ) have better 'mpg'
# (miles per gallon) values
```


### Styling options in boxplot

```python
# We can specify the order in which the different groups of the categorical 
# variable appears in the boxplot using 'order' parameter

sns.boxplot(x='origin',y='mpg',data=cars,order=['japan','europe','usa']);
```

```python
# color parameter
sns.boxplot(x='origin',y='mpg',data=cars,color='g');
```

```python
# width parameter

sns.boxplot(x='origin',y='mpg',data=cars,width=0.5); # default value of width=0.8
```

```python
# whis parameter -- default value for this is 1.5 but we can change it using 'whis' parameter 

sns.boxplot(x='origin',y='mpg',data=cars,whis=2);
# So we can see that by changing the whis value outliers have also changed for the groups
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,whis=1);
```

```python
# fliersize parameter-- It is used to change the size of the diamond points (outliers) -> default value is 5

sns.boxplot(x='origin',y='mpg',data=cars,fliersize=10);
```
