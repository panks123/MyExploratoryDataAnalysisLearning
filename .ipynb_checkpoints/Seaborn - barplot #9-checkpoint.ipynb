# Seaborn Barplot

**It is a categoriacal estimate plot. Show point estimates and confidence intervals as rectangular bars.A barplot is basically used to aggregate the categorical data according to some methods and by default itâ€™s the mean.To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset('penguins')
df.head()
```

```python
df.species.value_counts()
```

```python
df.dropna(inplace=True,how='any')
df.isnull().sum()
```

```python
sns.barplot(x='species',y='body_mass_g',data=df) # here 'species' is a categorical variable and 'body_mass_g' is numerical 
plt.show()
```

```python
# It has grouped the birds according to the 'species' column and then it has taken the mean of 'body_mass_g' of each group

# It can be seen as 
```

```python
df.groupby('species').body_mass_g.mean() # So the above bar plot has plotted using this data
```

```python
# Horizontal barplot --  this is obtained by interchanging the x and y axis variables as numerical and categorical

sns.barplot(y='species',x='body_mass_g',data=df)  
plt.show()
```

```python
# Confidence interval in the bar plot is the balck line on top of each bar
# The default value is 95 (can be interpreted as 95%)

sns.barplot(x='species',y='body_mass_g',data=df,ci=95) 
plt.show()
```

```python
sns.barplot(x='species',y='body_mass_g',data=df,ci=60) # we can see that we are getting a smaller black line on top of each bars
plt.show()
```

```python
# We can hide them as well
sns.barplot(x='species',y='body_mass_g',data=df,ci=None) 
plt.show()
```

```python
# Till now the bar plot is plotted by finding the mean of each group of the variable on x -axis 
# but we can change it by 'estimator' parameter e.g. we can pass it as standard deviation

sns.barplot(x='species',y='body_mass_g',data=df,estimator=np.std) 
plt.show()
```

```python
# It has grouped the birds according to the 'species' column and then it has taken the standard deviation
# of 'body_mass_g' of each group

# It can be seen as 
df.groupby('species').body_mass_g.std()
```

```python
# We can also use other methods in 'estimator' parameter (these methods can be any statistical method) e.g.

sns.barplot(x='species',y='body_mass_g',data=df,estimator=np.max) 
plt.show()
```

```python
df.groupby('species').body_mass_g.max()
```

```python
# hue parameter

# Till now we have used one categorical variable in the bar plot
#  Now we will use two
sns.barplot(x='species',y='body_mass_g',data=df,hue='sex') 
plt.show()
```

```python
# More styling in barplot
```

```python
# color -- specifies one single color for each bars
sns.barplot(x='species',y='body_mass_g',data=df,color='g') 
plt.show()
```

```python
# 'palette' specifies differnt color pattern for the bars of the bar plot

sns.barplot(x='species',y='body_mass_g',data=df,palette='dark') 
plt.show()
```

```python
# We can also change the color and the width of the ci line that appears on the top of the bars

sns.barplot(x='species',y='body_mass_g',data=df,color='g',errcolor='r',errwidth=10) 
plt.show()
```

```python
# edgecolor and lw parameters
sns.barplot(x='species',y='body_mass_g',data=df,color='g',errcolor='r',errwidth=10,edgecolor='k',lw=3) 
plt.show()
```
