# Seaborn pairplot
**A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. It is used to plot multiple pairwise bivariate distributions in a dataset**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.__version__
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# So we have three numerical variables and 4 categorical variable

# In the pairplot , the pairing of only numerical variables is done 
```

```python
sns.set_style('darkgrid')
```

```python
sns.pairplot(df);
```

```python
# So along the diagonal pairing is done with the same variable so along the diagonal we get some kind of distribution plot
# while in others pairing is done with different vriables so we get the relationship plot (scatter plot) for them
```

**We can plot differnt types of pairplot:**

```python
sns.pairplot(df,kind='kde');
```

```python
sns.pairplot(df,kind='reg');
```

```python
# Using 'hue' parameter with categorical vaiables
```

```python
df['time'].value_counts()
```

```python
sns.pairplot(df,hue='time');
```

```python
# Creating a column 'weakend' using the existing 'day' column
```

```python
df['weakend']=df.day.isin(['Sat','Sun'])
df.head()
```

```python
df.weakend.value_counts()
```

```python
sns.pairplot(df,hue='weakend');
```

```python
# If we don't want to see all the variables in the pairplot,we specify only those variales which we want in the 'vars' parameter

sns.pairplot(df,hue='weakend',vars=['total_bill','tip']);
```

```python
# We can also specify that which variable we want to see in the x-axis and which we want in the y-axis

sns.pairplot(df,hue='weakend',y_vars=['total_bill','tip'],x_vars=['total_bill','tip','size']);
```

```python
df.drop('weakend',axis=1,inplace=True)
```

```python
# We can style our pairplot by exclusively specifying keyword arguments as dictionary for diag_kws (diagonals), 
# plot_kws (off diagonals)

sns.pairplot(df,diag_kws={'color':'g'});
plt.show()
```

```python

sns.pairplot(df,kind='reg',hue='time',diag_kws={'color':'g'},plot_kws={'ci':False});
plt.show()
```
