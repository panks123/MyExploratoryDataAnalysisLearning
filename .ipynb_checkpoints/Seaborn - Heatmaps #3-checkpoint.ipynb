# Seaborn Heatmaps
**A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors. It is a graphical representation of data using colors to visualize the value of the matrix.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
arr_2d=np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)

plt.show()
```

```python
df=pd.read_csv('Who_is_responsible_for_global_warming.csv')
df
```

```python
df.drop(['Country Code','Indicator Name','Indicator Code'],axis=1,inplace=True)
df
```

```python
df.set_index('Country Name',inplace=True)
df
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df)
plt.show()
```

```python
# From the above heatmap we can derive that initilly United States had higher Co2 emission level but in the upcoming years it might have 
# taken some measures to control it as we can see the CO2 emission level is decreasing over years for united states

# But in China the Co2 emission is incresing over the years

#also we can see that for India. Pakistan, Nepal and Bangladesh CO2 emission level are minimal
```

```python
# vmin and vmax parameters -- If we set the vmin value to 5 and the vmax value to 21, 
# then only the cells with values between 5 and  21 will be displayed.

plt.figure(figsize=(12,7))

vmin=5
vmax=21

sns.heatmap(df,vmin=vmin,vmax=vmax)
plt.show()
```

```python
#cmap (color map) parameter -- to change the color map of the heatmap

plt.figure(figsize=(12,7))
sns.heatmap(df,cmap = "tab20")
plt.show()
```

```python
# annot (annotation) parameter -- to display the cell values

plt.figure(figsize=(12,7))
sns.heatmap(df,cmap = "tab20",annot=True)
plt.xlabel('Years')
plt.show()
```

```python
# We can also provide Our own annotations 

# e.g.

arr_2d=np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
annot_arr=np.array([
    ['a00','a01','a02'] ,
    ['a10','a11','a12'] ,
    ['a20','a21','a22'] ,
    ['a30','a31','a32'] 
])
annot_arr
```

```python
sns.heatmap(arr_2d,annot=annot_arr,fmt='s') # as our annotation cell values were string type so we passed fmt='s' 
# (fmt means format code)

plt.show()
```

```python
# linewidth and linecolor parameters -- for separating the cells of the heatmap
plt.figure(figsize=(12,7))
sns.heatmap(df,cmap = "tab20",annot=True,linewidth=1,linecolor='k') # black colored line of width 1 separates ech cell
plt.xlabel('Years')
plt.show()
```

```python
# cbar parameter - for Hiding the colorbar by maming it False

plt.figure(figsize=(12,7))
sns.heatmap(df,cmap = "tab20",annot=True,linewidth=1,linecolor='k',cbar=False) 
plt.xlabel('Years')
plt.show()
```

```python
# customizing the color bar
# cbar_kws (colorbar keywords) parameter

cbar_kws={'orientation':'horizontal','ticks':np.arange(0,22),'drawedges':True}

plt.figure(figsize=(12,7))
sns.heatmap(df,cmap = "tab20",annot=True,linewidth=1,linecolor='k', cbar_kws=cbar_kws) 
plt.xlabel('Years')
plt.show()
```

```python
cbar_kws={'orientation':'horizontal','shrink':1,'extend':'min', 'ticks':np.arange(0,22),'drawedges':True}

plt.figure(figsize=(12,7))
sns.heatmap(df,cbar=True,cbar_kws=cbar_kws)
plt.show()
```

# Correlation with heatmaps

**What is Correlation?**

It is the statistical measure of the relationship between two variables

It is a value between -1 to 1

- -ve correlation means the two variables have -ve relationship i.e. Increse in one variable decreases the other
- +ve correlation means +ve relationship i.e. with increase in on e variable other also increases
- 0 correlation means no relationship between two variables i.e with increse in one variable there is no effect in the other


```python
df=pd.read_csv('Tips.csv') # seaborn provides some inbuilt datasets
df.head()
```

```python
# Here in this dataset we can clearly see that 'TotalBill' is the dependent variable which is dependent on other variables
```

**Correlation heatmaps uses colored cells to show a 2D matrix correlation (table) between two discrete dimentions**

**It is used in feature selection**

```python
df.corr()  # it gives Pearson correlation i.e. df.corr(method='pearson') ->default other methods are also there 
#            like -->  method : {'pearson', 'kendall', 'spearman'}
# It finds the interrelationships between numerical variables
```

```python
# plotting the correlation heatmap
sns.heatmap(df.corr())
plt.show()
```

```python
# This is generally used in feature selection in machine learning e.g. suppose the dependent variable is 
# dependent on two features having same or almost same correlations then instead of choosing both the 
# features for training we can choose only one
```

```python
sns.heatmap(df.corr(),annot=True,linewidth=1,linecolor='k')
plt.show()
```

```python
# As the correlation matrix is symmetric 
# So we get  symmetric heatmap
```

```python
# plotting the triangular correlation heatmap

# creating mask
mask = np.triu(np.ones_like(df.corr()))
mask
```

```python
sns.heatmap(df.corr(),annot=True,linewidth=1,linecolor='k',mask=mask)
plt.show()
```

```python
sns.heatmap(df.corr(),cmap="YlGnBu",annot=True,mask=mask)
plt.show()
```
